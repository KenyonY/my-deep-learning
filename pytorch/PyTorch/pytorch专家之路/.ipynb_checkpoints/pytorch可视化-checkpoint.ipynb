### tensorboard中各个功能函数  

`torch.utils.tensorboard.writer.SummaryWriter（log_dir = None，comment =''，** kwargs ）`  
该SummaryWriter类提供了一个高层次的API来创建指定目录的事件文件，并添加摘要和事件给它。该类异步更新文件内容。  
**这允许训练程序调用方法直接从训练循环向文件添加数据，而不会减慢训练速度。**
* `add_scalar(tag, scalar_value, global_step=None, walltime=None)`  
在一个图表中记录一个标量的变化，常用于 Loss 和 Accuracy 曲线的记录。  
**参数：**  
tag（string） - 数据标签，即title  
scalar_value（float或string / blobname） - 要保存的值，对应y坐标  
global_step（int） - 要记录的全局步长值，对应x坐标    
walltime（float） - 可选覆盖默认的walltime（time.time（））


* `add_image(tag, img_tensor, global_step=None, walltime=None, dataformats='CHW')`  
Add image data to summary.  
绘制图片，可用于检查模型的输入，监测 feature map 的变化，或是观察 weight。


* `add_histogram(tag, values, global_step=None, bins='tensorflow', walltime=None, max_bins=None)`   
绘制直方图和多分位数折线图，常用于监测权值及梯度的分布变化情况，便于诊断网
络更新方向是否正确。  
参数：  
bins(string)- 决定如何取 bins，默认为‘tensorflow’，可选：’auto’, ‘fd’等
