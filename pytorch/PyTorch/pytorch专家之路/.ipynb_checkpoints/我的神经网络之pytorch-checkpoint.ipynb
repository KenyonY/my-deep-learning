## 建立最简单的神经网络: 
### 用两层神经网络来拟合函数：

```python
import torch
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
x = torch.unsqueeze(torch.linspace(0, 10, 100), dim=1)  # x data (tensor), shape=(100, 1)
y = torch.sin(x) #+ 0.2*torch.rand(x.size())                 # noisy y data (tensor), shape=(100, 1)

plt.scatter(x.data.numpy(), y.data.numpy())
plt.show()
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class Net(nn.Module):
    def __init__(self,n_input,n_hidden1,n_hidden2,n_output):
        super(Net,self).__init__()
        # 这里我定义了两个隐藏层 hidden1 hidden2
        self.hidden1 = nn.Linear(n_input, n_hidden1)
        self.hidden2 = nn.Linear(n_hidden1,n_hidden2)
        self.predict = nn.Linear(n_hidden2,n_output)
    def forward(self,x):
#         x = self.hidden(x)
############################################################
        x = F.relu(self.hidden1(x))
        x = F.sigmoid(self.hidden2(x))
############################################################
        x = self.predict(x)
        return x
```

```python
net = Net(1,5,5,1) # 搭建一个输入参数为1个，2个隐藏层中各有20个神经元，输出为1个的神经网络
print(net)

optimizer = torch.optim.SGD(net.parameters(),lr=0.1)
loss_func = nn.MSELoss() # 处理拟合时，可以用MSE,均方根误差
```

```python
plt.ion()
for t in range(2000):
    # 每一次循环将改变网络当前权重， 和梯度
    
    prediction = net(x)            # 计算网络当前权重下的的预测值
    
    loss = loss_func(prediction, y)# 计算预测值与实际值的误差
    
    optimizer.zero_grad()          # 将梯度至于0，否则会累加
                                   
                                  
    
    loss.backward()                # 计算各权重在当前位置梯度

    optimizer.step()               # 梯度下降
    
    
    if t % 200 == 0: # 绘图
        # plot and show learning process
        plt.cla()
        plt.scatter(x.data.numpy(), y.data.numpy())
        plt.plot(x.data.numpy(), prediction.data.numpy(), 'r-', lw=5)

        plt.text(0.5, 0, 'Loss=%.4f' % loss.item(), fontdict={'size': 20, 'color':  'red'})
        plt.show()
        plt.pause(0.2)

plt.ioff()
plt.show()
```

###  分类

```python
# 生成虚假数据
n_data = torch.ones(100, 2)
''' torch.normal(mean, std, out=None) -> Tensor '''
x0 = torch.normal(2*n_data, 1.2)      # class0 x data (tensor), shape=(100, 2)
y0 = torch.zeros(100)               # class0 y data (tensor), shape=(100, 1)
x1 = torch.normal(-2*n_data, 1.2)     # class1 x data (tensor), shape=(100, 2)
y1 = torch.ones(100)                # class1 y data (tensor), shape=(100, 1)

# cat 拼接
x = torch.cat((x0, x1), 0).type(torch.FloatTensor)  # shape (200, 2) FloatTensor = 32-bit floating
y = torch.cat((y0, y1), ).type(torch.LongTensor)    # shape (200,) LongTensor = 64-bit integer


plt.scatter(x.data.numpy()[:, 0], x.data.numpy()[:, 1], c=y.data.numpy(), s=100, lw=0, cmap='RdYlGn')
plt.show()
```

**用前面拟合的使用的网络样式**

```python
class Net(nn.Module):
    def __init__(self,n_input,n_hidden1,n_hidden2,n_output):
        super(Net,self).__init__() 
        
        # 这里我定义了两个隐藏层 hidden1 hidden2
        self.hidden1 = nn.Linear(n_input, n_hidden1)
        self.hidden2 = nn.Linear(n_hidden1,n_hidden2)
        self.predict = nn.Linear(n_hidden2,n_output)
    def forward(self,x):
#         x = self.hidden(x)
########################################################
        x = F.relu(self.hidden1(x))
        x = F.sigmoid(self.hidden2(x))
########################################################
        x = self.predict(x)
        return x
```

```python
net = Net(n_input=2, n_hidden1=10,n_hidden2=10, n_output=2)     # define the network
print(net)  # net architecture

# Loss and Optimizer
# Softmax is internally computed.
# Set parameters to be updated.
optimizer = torch.optim.ASGD(net.parameters(), lr=0.02)
loss_func = torch.nn.CrossEntropyLoss()  # the target label is NOT an one-hotted
 
```

`output=2`,表示二分类
如 
输出结果为[0,1]表示属于第二类  
输出结果为[1,0]表示属于第一类  
lossFunction选择的是`CrossEntropyLoss`,即 交叉熵  
使用到了softMax函数
它可以对应多分类，它的输出存在两个结果，一个是浮点数，表示的意义是概率，另一个是标签，就是显示分在哪一类  
如    
输出结果为[0.1,0.3,0.6]表示属于以0.6的概率会给到第三类   
也就是标签结果为[0,0,1]  

>$$
\text{Softmax}(x_{i}) = \frac{\exp(x_i)}{\sum_j \exp({x_j})}$$

```python
plt.ion()   # something about plotting

for t in range(100):
    out = net(x) # 如果是二分类，out每个输出对应两个，分别表示在一类还是二类，经过softmax之后意义转为概率
    loss = loss_func(out, y)     

    optimizer.zero_grad()   
    loss.backward()         
    optimizer.step()        
    
    if t % 10 == 0 or t in [3, 6]:
        # plot and show learning process
        plt.cla()
        _, prediction = torch.max(F.softmax(out,dim=1), 1) # torch.max(input,dim)
        pred_y = prediction.data.numpy().squeeze()
        target_y = y.data.numpy()
        plt.scatter(x.data.numpy()[:, 0], x.data.numpy()[:, 1], c=pred_y, s=100, lw=0, cmap='RdYlGn')
        accuracy = len(pred_y[pred_y == target_y])/len(pred_y)
        plt.text(1.5, -4, 'Accuracy=%.2f' % accuracy, fontdict={'size': 20, 'color':  'red'})
        plt.show()
        plt.pause(0.1)

plt.ioff()
```

### 批训练 batch

当数据量非常大时，我们需要将这一堆数据分成多个一小批一小批的数据。

```python
import torch
import torch.utils.data as Data

BATCH_SIZE = 5 # 每一批有多少个数据
x = torch.linspace(1,10,10)
y = torch.linspace(10,1,10)

dataset = Data.TensorDataset(x, y) # 定义需要进行批训练的数据集

# 定义需要加载进去的模式
loader = Data.DataLoader( 
    dataset=dataset,
    batch_size=BATCH_SIZE,
    shuffle = True # 打乱小批数据的顺序
# num_workers (int, optional): 这个参数决定了有几个进程来处理data loading。0意味着所有的数据都会被load进主进程。（默认为0）
    
)

for epoch in range(3):# 整个数据集训练3次
    for step, (batch_x,batch_y) in enumerate(loader): # 在需要同时用到 index 和 value时，我们用enumerate()
        
        '''然后整个数据集被拆分成N个小批数据集，每个小批数据大小是BATCH_SIZE个，并且在前面的loader的定义中，
        我们对每个小批数据进行了打乱顺序的操作，每一大批训练的结果是不同的，这样相当于交叉验证'''
        
        print('Epoch: ', epoch, '| Step: ', step, '| batch x: ',
              batch_x.numpy(), '| batch y: ', batch_y.numpy())
```

## 卷积神经网络

```python
import torch
import torch.nn as nn
import torch.utils.data as Data
import torchvision
import matplotlib.pyplot as plt

torch.manual_seed(1)    # reproducible
```

```python
# Hyper Parameters
EPOCH = 1               # train the training data n times, to save time, we just train 1 epoch
BATCH_SIZE = 50
LR = 0.001              # learning rate
DOWNLOAD_MNIST = False   # set to False if you have downloaded
```

```python
# Mnist digits dataset
# 下面这个语句会让torch去Mnist下载文件到root文件夹，可通过shift+tab查看
# 当然 这里已经下载好了，直接到指定目录打开就好了
train_data = torchvision.datasets.MNIST(
    root='.\pytorch Folder\pytorch by 莫凡\mnist/',
    train=True,                                     # this is training data
    transform=torchvision.transforms.ToTensor(),    # torchvision.transforms 会将图片的rgb范围从[0,255] 压缩至[0,1]
                                                    # torch.FloatTensor of shape (C x H x W) and normalize in the range [0.0, 1.0]
    download=DOWNLOAD_MNIST,                        # download it if you don't have it
)
```

```python
# plot one example
print(train_data.data.size())                 # (60000, 28, 28)
print(train_data.targets.size())               # (60000)
# take a look at index = 5
_ = 50000
plt.imshow(train_data.data[_].numpy(), cmap='gray')
plt.title('%i' % train_data.targets[_])
plt.show()
```

```python

train_loader = Data.DataLoader( 
    dataset=train_data,
    batch_size=BATCH_SIZE,
    shuffle = True, # 打乱小批数据的顺序
    num_workers =2 # 数据集中有几个成员，本例中有 x ,y 所以是两个
    
)

test_data = torchvision.datasets.MNIST(
    root='.\pytorch Folder\pytorch by 莫凡\mnist/',
    train=False,# 这里为False时，表示数据集为test_data的
)
print(test_data.data.size()) # 我们取前2000个test数据集
test_x =torch.unsqueeze(test_data.data, dim=1).type(torch.FloatTensor)[:2000]/255.   # shape from (2000, 28, 28) to (2000, 1, 28, 28), value in range(0,1)
test_y = test_data.targets[:2000]
```

```python
class CNN(nn.Module):
    def __init__(self):
        super(CNN,self).__init__()
        self.conv1 = nn.Sequential(
            # 我们图片的维度此时是 （1*28*28）
            # 定义过滤器属性：
            nn.Conv2d(
                in_channels = 1, # 输入层的数据高度，这里是黑白数据，所以高度为1，若是rgb，则为3
                out_channels=16, # 输出到下一层的高度，
                kernel_size = 5, # 卷积的区域像素，这里表示对5*5的长宽的矩形区域进行卷积
                stride = 1,   # 卷积滤波区域以多少步长进行扫描
                padding = 2, # 滤波器的区域扫描到最后会超出原始数据像素的范围，所以在像素的周围用2圈0给它围住
                            #  如果stride=1, 则padding = (kernel_size-1)/2 = (5-1)/2
            ),# -> (16,28,28) 这一步，高度和宽度是不变的
            
            nn.ReLU(),# -> (16,28,28)
            nn.MaxPool2d(kernel_size=2) # 相当于再进行一次卷积，但高度不变
            # -> (16,14,14)
        )
        self.conv2 = nn.Sequential(# -> (16,14,14)
            nn.Conv2d(16,32,5,1,2),# -> (32,14,14)
            nn.ReLU(),# -> (32,14,14)
            nn.MaxPool2d(2)# -> (32,7,7)
            
        )
        self.out = nn.Linear(32*7*7,10) # nn.Linear(in_features, out_features, bias=True)
        # 这里 32*7*7表示最后一层是全连接层
        # 这里没有在最后一层加入softmax激活层
        
    def forward(self,x):
        x1 = self.conv1(x)
        x2 = self.conv2(x1) # 它的维度是(batch, 32,7,7),batch是指小批数据的维度
        x2 = x2.view(x2.size(0), -1) # (batch, 32*7*7)
        output = self.out(x2)
        
        return output,x2
```

```python
cnn = CNN()
print(cnn)
```

```python
# 训练
optimizer = torch.optim.Adam(cnn.parameters(),lr=0.0015)
loss_func = nn.CrossEntropyLoss()
```

```python
# training and testing
plt.ion
for epoch in range(EPOCH):
    for step, (x, y) in enumerate(train_loader):   # gives batch data, normalize x when iterate train_loader
        b_x = x   # batch x
        b_y = y   # batch y
        
#         print(b_x.size())
        output = cnn(b_x)[0]              # cnn output

        loss = loss_func(output, b_y)   # cross entropy loss
        
        optimizer.zero_grad()           # clear gradients for this training step
        loss.backward()                 # backpropagation, compute gradients
        optimizer.step()                # apply gradients
        
        if step % 100 == 0:# 每100步，我们拿当前的网络与test数据集进行测试
            test_output, last_layer = cnn(test_x)
            print(test_output.data.size())
            pred_y = torch.max(test_output, 1)[1].data.squeeze()
            accuracy = len(test_y[test_y==pred_y]) / float(test_y.size(0))
            print('Epoch: ', epoch, '| train loss: %.4f' % loss.item(), '| test accuracy: %.2f' % accuracy)
#             if HAS_SK:
#                 # Visualization of trained flatten layer (T-SNE)
#                 tsne = TSNE(perplexity=30, n_components=2, init='pca', n_iter=5000)
#                 plot_only = 500
#                 low_dim_embs = tsne.fit_transform(last_layer.data.numpy()[:plot_only, :])
#                 labels = test_y.numpy()[:plot_only]
#                 plot_with_labels(low_dim_embs, labels)
plt.ioff()  
```
