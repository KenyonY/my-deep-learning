```python
import torch
import numpy
from time import time
import matplotlib.pyplot as plt
torch.set_default_tensor_type('torch.cuda.FloatTensor')
```

```python
# gpu --> cpu
def g2c(N,epochs=100):
    shape = N,N
    m_gpu = torch.ones(shape)* 0.5
    
    t0 = time()
    for i in range(epochs):
        s = m_gpu.cpu()
    dt = (time()-t0)*1e3
    return dt/epochs

# gpu --> cpu
def c2g(N,epochs=100):
    shape = N,N
    m_cpu = np.ones(shape) * 0.5

    t0 = time()
    for i in range(epochs):
        s = torch.tensor(m_cpu)
    dt = (time()-t0)*1e3
    return dt/epochs

epochs = 50
t_g2c = []
t_c2g = []
L = 45
for i in range(L):
    N = i*100 + 512
    t_g2c.append(g2c(N, epochs))
    t_c2g.append(c2g(N, epochs))
```

```python
t_g2c = np.array(t_g2c)
t_c2g = np.array(t_c2g)
x = 512 + 100*np.arange(L)
plt.plot(x, t_g2c)
plt.plot(x, t_c2g)
plt.plot(x, t_g2c+t_c2g)
plt.legend(['gpu->cpu', 'cpu->gpu', 'total'])
plt.xlabel('shape.w and shape.h')
plt.ylabel('/ms')
```

# 矩阵运算GPU的提速效率


```python
import cv2

kernel = np.array([[0,  1, 0],
                   [1, -4, 1],
                   [0,  1, 0]], dtype=np.float32)

N=10
# cpu
def test_cpu(N,kernel):
    shape = N,N
    m = np.ones(shape,dtype=np.float32)* 0.5
    res = cv2.filter2D(m, -1, kernel,borderType=0)
    
# # gpu
def test_gpu(N,kernel):
    shape = N,N
    
    kernel = torch.tensor(kernel)[None, None,...]
    m = torch.ones(shape,dtype=torch.float32)* 0.5
    res = torch.conv2d(m[None,None,...], kernel,padding=1)
    
test_cpu(N, kernel)
test_gpu(N, kernel)
```

```python
import cv2

    
def c(N, epochs=100):
    shape = N,N
    m_cpu = np.ones(shape,dtype=np.float32)* 0.5
    t0 = time()
    for i in range(epochs):
        test_cpu(N, kernel)
    dt = (time()-t0)*1e3
    return dt/epochs

#gpu
def g(N, epochs=100):
    shape=N,N
    m_gpu = torch.ones(shape,dtype=torch.float32)* 0.5
    
    t0 = time()
    for i in range(epochs):
        test_gpu(N,kernel)
    dt = (time()-t0)*1e3
    return dt/epochs

torch.set_default_tensor_type('torch.FloatTensor')
epochs = 50
t_c = []
t_g = []
L = 20
for i in range(L):
    N = i*100 + 512
    t_c.append(c(N, epochs))
    t_g.append(g(N, epochs))
```

```python
t_c = np.array(t_c)
t_g = np.array(t_g)
x = 512 + 100*np.arange(L)
plt.plot(x, t_c)
plt.plot(x, t_g)

plt.legend(['cpu', 'torch.cpu'])
plt.xlabel('shape.w and shape.h')
plt.ylabel('/ms')
```
