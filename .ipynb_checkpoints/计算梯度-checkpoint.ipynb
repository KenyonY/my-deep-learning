# 计算多维数组的数值梯度

```python
def eval_numerical_gradient(f, x):
    '''
    - x a numpy vector
    - f give the map f(x)
    - return the gradient of f(x)
    '''

    grad = np.zeros(x.shape)
    dx = 1e-6
    
    # Calculate the gradient of each dimension
    it = np.nditer(x, flags = ['multi_index'], op_flags=['readwrite'])
    while not it.finished:
        ix = it.multi_index
        fx = f(x[ix])
        fxh = f(x[ix]+ dx)
        
        # culculate gradient
        grad[ix] = (fxh - fx) / dx
        it.iternext()
    return grad
```

```python
x = np.linspace(np.pi,2*np.pi, 100)
f=lambda x: np.sin(x)
plt.plot(x, eval_numerical_gradient(f,x))
```
